```python
import re

def parse_amount_indian(value):
    if pd.isna(value):
        return 0
    s = str(value).strip().lower()
    if s in {"no", "na", "n/a", "-", "—", "none", ""}:
        return 0
    s = (s.replace("₹","").replace("rs.","").replace("rs","")
           .replace("inr","").replace("/-","").replace(" ","").replace(",",""))

    # pure number
    if re.fullmatch(r"\d+(?:\.\d+)?", s):
        return int(float(s))

    # units
    m = re.fullmatch(r"(\d+(?:\.\d+)?)(l|lac|lakh|lakhs)", s)
    if m: return int(float(m.group(1))*100000)

    m = re.fullmatch(r"(\d+(?:\.\d+)?)(cr|crore|crores)", s)
    if m: return int(float(m.group(1))*10000000)

    m = re.fullmatch(r"(\d+(?:\.\d+)?)(k)", s)
    if m: return int(float(m.group(1))*1000)

    # fallback
    m = re.search(r"(\d+(?:\.\d+)?)", s)
    if m:
        num = float(m.group(1))
        if any(u in s for u in ["lakh","lakhs","lac","l"]): return int(num*100000)
        if any(u in s for u in ["cr","crore","crores"]):     return int(num*10000000)
        if "k" in s:                                        return int(num*1000)
        return int(num)
    return 0
```

```python
dataset["Security_deposit"]= dataset["Security_deposit"].apply(parse_amount_indian)
```

```python
dataset["Rent_price"]= dataset["Rent_price"].apply(parse_amount_indian)
```
