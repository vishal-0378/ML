```python
#importing pandas for data manipulation like read i used
import pandas as pd
```

```python
#to read the data from the housing amount dataset
df=pd.read_csv("/content/sample_data/california_housing_train.csv")
df.head()
```

```python
#here the x as feature and y as predicted output
x = df.drop('median_house_value', axis=1)
y = df['median_house_value']
```

```python
#train test split
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2,random_state=42)
```

```python
#import sslr libr
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
#import algo (LINEAR REGRESSION)
from sklearn.svm import SVR
```

```python
modle = SVR()
```

```python
#model fit
modle.fit(x_train, y_train)
```

```python
#train the LR model
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
#model prediction
y_pred = modle.predict(x_test)
```

```python
#accuracy score
modle.score(x_test,y_test)
```

```python
#to save the modle by using pickle
import pickle
pickle.dump(modle,open('modle.pkl','wb'))
pickle.dump(scaler,open('scaler.sav','wb'))
```
